## Question-2 :Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
df=sns.load_dataset("fmri")
```

```python
df.head()
```

```python
x=df["timepoint"]
y=df["signal"]
```

```python
sns.lineplot(x=df["timepoint"],y=df["signal"],data=df)
```

## Question-3 :Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
df1=sns.load_dataset("titanic")
```

```python
sns.boxplot(x=df1['pclass'],y=df1['age'],data=df1)
```

```python
sns.boxplot(x=df1["pclass"],y=df1["fare"],data=df1)
```

## Question-5 :Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
df4=sns.load_dataset("iris")
```

```python
sns.relplot(x=df4["sepal_length"],y=df4["sepal_width"],hue=df4["species"])
```
