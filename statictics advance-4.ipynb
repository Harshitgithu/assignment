## Question-1 :What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

```python
Both t-tests and z-tests are statistical methods used to make inferences about population parameters based on sample data. The main difference between the two lies in the circumstances under which they are appropriate.

1. T-test:

Use Case: The t-test is typically employed when the sample size is small (usually less than 30) or when the population standard deviation is unknown.
Example Scenario: Suppose you want to compare the mean heights of two groups of students (Group A and Group B). You collect a sample from each group, and since you don't know the population standard deviation, you use a t-test to determine if there is a significant difference in the mean heights between the two groups.
2. Z-test:

Use Case: The z-test is suitable when dealing with large sample sizes (usually greater than 30) and when the population standard deviation is known.
Example Scenario: Imagine you are studying the average test scores of all high school students in a particular city. You have a large dataset, and you know the population standard deviation. In this case, you can use a z-test to assess whether the mean test score of the entire population is significantly different from a specified value.
In summary, the choice between a t-test and a z-test depends on the sample size and whether the population standard deviation is known. If the sample size is small or the population standard deviation is unknown, a t-test is appropriate. For large sample sizes with a known population standard deviation, a z-test is more suitable.
```

## Question-2 :Differentiate between one-tailed and two-tailed tests.

```python
Definition: A one-tailed test is a statistical hypothesis test where the critical region of the distribution is on only one side (either the right or the left) of the probability distribution.
Use Case: It is typically used when there is a specific directional hypothesis, and you are interested in testing whether the sample statistic is significantly greater than or less than a specified value.
Example: Suppose you are conducting a study to test whether a new drug increases the average recovery time of patients. Your one-tailed hypotheses would be:
Null Hypothesis (H0): The new drug has no effect on recovery time.
Alternative Hypothesis (H1): The new drug decreases recovery time (left-tailed) or increases recovery time (right-tailed).
Two-Tailed Test:

Definition: A two-tailed test is a statistical hypothesis test where the critical region is on both sides of the probability distribution. It is used to test if a sample statistic is significantly different from a specified value.
Use Case: It is employed when there is no specific directional hypothesis or when you want to detect if a parameter is significantly different in either direction.
Example: Consider a scenario where you are investigating whether a new teaching method affects students' test scores. Your two-tailed hypotheses would be:
Null Hypothesis (H0): The new teaching method has no effect on test scores.
Alternative Hypothesis (H1): The new teaching method significantly affects test scores (in either direction).
In summary, the key distinction lies in the directionality of the hypothesis. One-tailed tests focus on a specific direction (either greater than or less than), while two-tailed tests consider the possibility of a significant difference in either direction from the specified value. The choice between them depends on the research question and the specific hypotheses being tested.




```

## Questio-3 :Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

```python
Type I Error:

Definition: A Type I error occurs when a true null hypothesis is incorrectly rejected. In other words, it is the mistake of concluding that there is a significant effect or difference when there is none in reality. The probability of committing a Type I error is denoted by the symbol α (alpha), which is the significance level of the test.
Example Scenario for Type I Error:
Suppose a pharmaceutical company is testing a new drug, and the null hypothesis (H0) is that the drug has no side effects. The alternative hypothesis (H1) is that the drug does have side effects. If the researchers mistakenly reject the null hypothesis and conclude that the drug has side effects when, in fact, it does not, it is a Type I error.

Type II Error:

Definition: A Type II error occurs when a false null hypothesis is not rejected. It is the mistake of failing to detect a significant effect or difference when one actually exists. The probability of committing a Type II error is denoted by the symbol β (beta).
Example Scenario for Type II Error:
Continuing with the drug example, let's say the drug does have side effects, but the researchers fail to reject the null hypothesis that there are no side effects. This would be a Type II error – the failure to detect a real effect.

In summary, Type I error involves incorrectly rejecting a true null hypothesis, leading to a false positive conclusion, while Type II error involves failing to reject a false null hypothesis, leading to a false negative conclusion. The balance between these two errors is often controlled by adjusting the significance level (α) and the power of the test, and it is crucial to consider both types of errors when interpreting the results of hypothesis tests.




```
